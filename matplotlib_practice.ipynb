```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with object-oriented method for more complex charts ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() another way
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() another way again
ax = plt.axes()
ax.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston')
ax.grid()
ax.set_title("PyBer Fare by Month")
ax.set_xlabel('Date')
ax.set_ylabel('Fares($)')
ax.set_ylim(0, 45)
ax.legend()
# To change figure-level attributes (such as axis labels, a title, or a legend) 
# or save the figure as an image, use fig = plt.figure().
```

```python
# Create the plot using plt.show - looks for all the active figure 
# objects and opens a window that displays them if you have two or 
# more sets of data to plot. CAUTION:
# Should only be used once per session, can cause the graph to display improperly
plt.plot(x_axis, y_axis)
#plt.show()
```

```python
# Create plot and label fo legend
plt.plot(x_axis, y_axis, label='Boston')

# Create labels for the axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

#Set the y limit between 0 and 45
plt.ylim(0, 45)

# Create a title
plt.title("Pyber Fare by Month")

# Add the legend
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="red", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```
