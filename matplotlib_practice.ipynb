# %matplotlib inline

```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

## Create the plot
plt.plot(x_axis, y_axis)

```python
# Create the plot with object-oriented method for more complex charts ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() another way
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() another way again
ax = plt.axes()
ax.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston')
ax.grid()
ax.set_title("PyBer Fare by Month")
ax.set_xlabel('Date')
ax.set_ylabel('Fares($)')
ax.set_ylim(0, 45)
ax.legend()
# To change figure-level attributes (such as axis labels, a title, or a legend) 
# or save the figure as an image, use fig = plt.figure().
```

```python
# Create plot and label fo legend
plt.plot(x_axis, y_axis, label='Boston')

# Create labels for the axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

#Set the y limit between 0 and 45
plt.ylim(0, 45)

# Create a title
plt.title("Pyber Fare by Month")

# Add the legend
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="red", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create a Horizontal Bar Chart
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, label='Chicago', color="cyan")
ax.legend()
ax.set_xlabel('Fares ($)')
ax.set_ylabel('Date')
plt.gca().invert_yaxis()
```

```python
# Create a Scatter Plot
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# add a lowercase "o" as a parameter to switch from line to scatter plot
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(y_axis, x_axis, c='red', label='Chicago')
plt.legend()
plt.xlabel('Fare ($)')
plt.ylabel('Date')
plt.title("PyBer Fare by Month")
plt.gca().invert_yaxis()
plt.xlim(0, 50)
```

```python
# Create a Bubble Chart
# plt.scatter(x_axis, y_axis, s=y_axis_larger)
# y_axis_larger = []
# for data in y_axis:
#  y_axis_larger.append(data*3)
```

```python
# use list comprehension to replace many for and while loops
# new_list = [expression for item in list if conditional]
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Create a Bubble Chart Using the Object-Oriented Interface
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, c = 'skyblue', label = 'Boston', s = [i * 5 for i in y_axis], alpha=0.8, edgecolors= "black", linewidth=2)
ax.set_title('PyBer Fare by Month')
plt.gca().invert_yaxis()
ax.set_xlim(0, 50)
ax.legend()
ax.set_xlabel('Fares ($)')
ax.set_ylabel('Date')
```

```python
# Create a Pie Chart Using the MATLAB Method
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

# Create a Pie Chart Using the Object-Oriented Interface

```python
fig, ax = plt.subplots(figsize=(8, 8))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
ax.pie(y_axis,labels=x_axis, shadow = True, colors=colors, explode=explode_values, startangle=95, autopct='%.1f%%', counterclock=False)
plt.show()
```
